```python
#importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
a=[]
b=[]
i=0
for page in range(0,205,5):
    c=requests.get("https://www.tripadvisor.in/Hotel_Review-g503703-d2038443-Reviews-or"+str(page)+"-Mahodadhi_Palace-Puri_Puri_District_Odisha.html")
    t=c.content
    soup=BeautifulSoup(t,"html.parser")
    right_table=soup.find("div",class_="listContainer hide-more-mobile")
    for row in right_table.findAll("div",class_="quote"):
        print("Scraping please wait",end='')
        cells=row.findAll("span",class_="noQuotes")
        a.append(i)
        b.append(cells[0].find(text=True))
        i=i+1
        print("...")
import pandas as pd
df=pd.DataFrame()
df['sl no.']=a
df['reviews']=b
print(df)
```

```python
df.to_csv('amit2nd.csv')
```
