```python
import requests
from bs4 import BeautifulSoup
```

```python
a=[]
b=[]
i=0
for page in range(0,200,5):
    c=requests.get("https://www.tripadvisor.in/Hotel_Review-g503703-d3949517-Reviews-or"+str(page)+"-Hotel_Pushpa-Puri_Puri_District_Odisha.html")
    t=c.content
    soup=BeautifulSoup(t,"html.parser")
    row_table=soup.find("div",class_="listContainer hide-more-mobile")
    for row in row_table.findAll("div",class_="quote"):
        print("Scraping please wait",end='')
        cells=row.findAll("span",class_="noQuotes")
        a.append(i)
        b.append(cells[0].find(text=True))
        i=i+1
        print("...")
import pandas as pd
df=pd.DataFrame()
df['sl.no']=a
df['Reviews']=b
print(df)
df.to_csv('amit3rd.csv')    
    
```
